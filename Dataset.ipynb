## **Medidas de Tendencia Central**

El análisis de tendencia central permite identificar los valores típicos o representativos del comportamiento de las variables relacionadas con el estrés académico y los hábitos estudiantiles. Para este estudio se calcularon la **media**, la **mediana** y la **moda** de todas las variables numéricas incluidas en el dataset.

### **Interpretación individual de las medidas:**

- **Media (promedio):**  
  Representa el valor promedio reportado por los estudiantes. En escalas tipo Likert (1 a 5), una media cercana a 3 corresponde a un nivel moderado, mientras que valores superiores a 3 indican presencia elevada de la característica evaluada (por ejemplo, alto estrés o alta carga académica).

- **Mediana:**  
  Indica el punto central de los datos. Es menos sensible a valores extremos y resulta útil para identificar la tendencia general sin influencia de casos atípicos. Una mediana diferente a la media sugiere que la distribución podría estar sesgada.

- **Moda:**  
  Representa el valor más frecuente dentro de la variable. En este estudio, su relevancia radica en la identificación del nivel más común reportado por los estudiantes en aspectos como estrés, rendimiento, calidad del sueño o carga académica.

### **Comparación entre media y mediana:**

La relación entre ambas medidas permite evaluar la forma de la distribución:

- **Media > Mediana:** indica un sesgo positivo (valores altos más frecuentes).  
- **Media < Mediana:** indica un sesgo negativo (valores bajos más frecuentes).  
- **Media ≈ Mediana:** sugiere simetría en la distribución.

Este contraste es fundamental para comprender si la percepción estudiantil está concentrada en valores bajos, moderados o altos.



# **Medidas de Dispersión**

```python
import pandas as pd
from tabulate import tabulate

# Columna que deseas analizar
columna = "Stress_Levels"

# Calcular medidas de dispersión
dispersion = pd.DataFrame({
    "Rango": [df[columna].max() - df[columna].min()],
    "Varianza": [df[columna].var()],
    "Desviación Estándar": [df[columna].std()],
    "Coeficiente de Variación (%)": [(df[columna].std() / df[columna].mean()) * 100],
    "Percentil 25 (Q1)": [df[columna].quantile(0.25)],
    "Percentil 75 (Q3)": [df[columna].quantile(0.75)],
    "IQR (Rango intercuartílico)": [df[columna].quantile(0.75) - df[columna].quantile(0.25)]
})

# Mostrar tabla en formato markdown
print(tabulate(dispersion.round(2), headers="keys", tablefmt="github"))
```

**Resumen de las Medidas de Dispersión**

Las medidas de dispersión permiten evaluar qué tan dispersos o concentrados se encuentran los valores de cada variable alrededor de su media. En este conjunto de datos, estas métricas proporcionan una visión complementaria al análisis de tendencia central, mostrando la variabilidad existente entre las respuestas de los estudiantes.

- **Desviación estándar (std):** Indica cuánto se alejan, en promedio, los valores de cada variable respecto a su media. Valores más altos reflejan mayor variabilidad en las respuestas.
En este dataset, variables como Niveles de Estrés y Actividades Extracurriculares presentan desviaciones relativamente amplias, lo que sugiere que las percepciones entre estudiantes son diversas.

- **Varianza:**
Representa la dispersión al cuadrado de los datos. Su utilidad es principalmente comparativa, ya que permite observar qué variables tienen mayor fluctuación interna.

- **Rango (max – min):** Mide la amplitud total de las respuestas. La mayoría de las variables oscilan entre 1 y 5, lo que indica que todos los estudiantes responden dentro de los valores posibles, pero el rango ayuda a confirmar si hay categorías usadas muy poco o muchísimo.

- **Coeficiente de variación (CV):** Expresa la dispersión relativa en función de la media. Permite comparar la variabilidad entre variables aunque tengan medias distintas.
Un CV alto indica que las respuestas son heterogéneas; un CV bajo sugiere consistencia entre los estudiantes.

En general, las medidas de dispersión confirman que, aunque los estudiantes comparten ciertos patrones comunes, existe una variabilidad significativa en aspectos como niveles de estrés, carga de estudio y descanso, lo que evidencia diferencias importantes en sus experiencias académicas y personales.

# **Límites para detección de outliers (Datos simétricamente distribuidos)**

Cuando los datos se consideran simétricamente distribuidos, se pueden detectar valores atípicos usando la regla basada en la desviación estándar (σ).
Un valor se considera outlier si se encuentra demasiado alejado de la media.

Para cada variable:
- Límite inferior= μ − 3σ
- Límite superior= 𝜇 + 3𝜎

Donde:

- μ = media

- σ = desviación estándar

- 3σ corresponde a la regla empírica (99.7% de los datos en una distribución normal)

```python
# Límites para detección de outliers asumiendo distribución simétrica

# Excluir columnas no numéricas antes de calcular media y desviación estándar
numeric_df = df.select_dtypes(include=['number'])

outlier_limits = pd.DataFrame({
    "Media": numeric_df.mean(),
    "Desviación estándar": numeric_df.std(),
})

outlier_limits["Límite inferior (μ - 3σ)"] = outlier_limits["Media"] - 3 * outlier_limits["Desviación estándar"]
outlier_limits["Límite superior (μ + 3σ)"] = outlier_limits["Media"] + 3 * outlier_limits["Desviación estándar"]

outlier_limits.round(2)
```
